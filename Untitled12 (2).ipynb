**Step 2:**Import Libraries:

**osmnx:** A Python library for working with OpenStreetMap data.
**folium:** A Python wrapper for Leaflet.js, a JavaScript library for interactive maps.

**Step 3:**Define the Area of Interest:

Set the variable place_name to the name of the city or area of interest (in this case, Bangkok).

**Step 4:**Fetch Features Data Using OSMnx:

Use OSMnx to fetch features data for the specified place, focusing on amenities with the tag 'hospital'.

**Step 5:**Create a Folium Map:

Create a Folium map centered at the specified location (latitude 13.7563, longitude 100.5018) with an initial zoom level of 13.

**Step 6:**Iterate Over Hospital Features and Add Markers to the Map:

Iterate over the rows of the features DataFrame.
For each hospital feature, extract its centroid coordinates, and name, and add a marker with a blue hospital icon to the Folium map.

**Step 7:**Display the Map in the Colab Notebook:

The last line, when executed in a Colab notebook, will display the Folium map.

```python
import osmnx as ox
import folium

# Define the city or area of interest (e.g., Bangkok)
place_name = "Bangkok"

# Fetch features data using OSMnx
features = ox.features_from_place(place_name, tags={'amenity': 'hospital'})

# Create a Folium map centered at the location of Bangkok
map_hospitals = folium.Map(location=[13.7563, 100.5018], zoom_start=13)

# Iterate over all hospital features and add them to the map with custom icons
for _, feature in features.iterrows():
    lat, lon = feature['geometry'].centroid.y, feature['geometry'].centroid.x
    name = feature['name'] if 'name' in feature else 'Hospital'
    folium.Marker([lat, lon], popup=name, icon=folium.Icon(icon='hospital', color='blue')).add_to(map_hospitals)

# Display the map in the Colab notebook
map_hospitals
```

## **Steps to visualize the number of hospitals attending emergency cases:**

**Step 1:**

`pip install geopandas` installs the Geopandas library, extending Pandas for geospatial data. It enables spatial operations and data structures for working with geographic information in Python.

```python
pip install geopandas
```

**Step 2:**Importing Libraries:

geopandas is a library for working with geospatial data, and matplotlib is used for data visualization, including plotting graphs and charts.

**Step 3:**Specify the Path to GeoJSON File:

geojson_path is a variable that holds the file path to the GeoJSON dataset.

**Step 4:**Read GeoJSON Dataset:

gpd.read_file reads the GeoJSON file and creates a GeoDataFrame (gdf), which is similar to a Pandas DataFrame but with additional geospatial capabilities.

**Step 5:**Count Hospitals:

Counts the total number of hospitals, emergency hospitals, and non-emergency hospitals in the GeoDataFrame.

**Step 6:**Display Counts:

Prints the counts of total, emergency, and non-emergency hospitals.

**Step 7:**Plotting:

Creates a bar chart to visualize the count of hospitals with and without emergency services

**Step 8:**Annotate the Bars:

Annotates the bars on the chart with the corresponding count.

**Step 9:**Display the Plot:

Shows the plot in the output.





```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Specify the path to your GeoJSON file
geojson_path = '/content/export (1).geojson'

# Use try-except block to handle potential errors
try:
    # Read the GeoJSON dataset
    gdf = gpd.read_file(geojson_path)

    # Count of all hospitals
    total_hospitals = len(gdf)

    # Count of emergency hospitals
    emergency_hospitals = gdf[gdf['emergency'] == 'yes'].shape[0]

    # Count of non-emergency hospitals
    non_emergency_hospitals = gdf[gdf['emergency'] == 'no'].shape[0]

    # Display the counts
    print(f"Total Hospitals: {total_hospitals}")
    print(f"Emergency Hospitals: {emergency_hospitals}")
    print(f"Non-emergency Hospitals: {non_emergency_hospitals}")

    # Plotting
    plt.figure(figsize=(8, 6))
    plt.bar(['Hospitals with Emergency', 'Hospitals without Emergency'], [emergency_hospitals, non_emergency_hospitals], color=['red', 'blue'])
    plt.xlabel('Type')
    plt.ylabel('Count')
    plt.title('Count of Hospitals with and without Emergency Services')

    # Annotate the bars with the count
    for i, bar in enumerate(plt.bar(['Hospitals with Emergency', 'Hospitals without Emergency'], [emergency_hospitals, non_emergency_hospitals], color=['red', 'blue'])):
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, height, f'{int(height)}', ha='center', va='bottom')

    # Display the plot
    plt.show()

except Exception as e:
    print(f"Error reading GeoJSON file: {e}")


```
